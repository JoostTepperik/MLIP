## Handling missing values

```python
def clean_data(df,test = 0):
    df = df.copy()
    
    # Identify truly numerical columns (checking data type and content)
    numerical_columns = [
        'donor_age', 'karnofsky_score', 'melphalan_dose',
        'comorbidity_score'
    ]
    
    # Columns to fill with mode (most frequent value)
    mode_fill_columns = [
        'psych_disturb', 'diabetes', 'vent_hist', 'cmv_status',
        'in_vivo_tcd', 'gvhd_proph', 'sex_match', 'donor_related',
        'dri_score', 'cyto_score'  # Moved from numerical to mode filling
    ]
    
    # All other categorical columns
    other_categorical_columns = [
        'hla_match_c_high', 'hla_high_res_8', 'arrhythmia', 'hla_low_res_6',
        'renal_issue', 'pulm_severe', 'hla_high_res_6', 'tce_imm_match',
        'hla_nmdp_6', 'hla_match_c_low', 'rituximab', 'hla_match_drb1_low',
        'hla_match_dqb1_low', 'cyto_score_detail', 'hla_high_res_10',
        'hla_match_dqb1_high', 'conditioning_intensity', 'ethnicity',
        'obesity', 'mrd_hct', 'tce_match', 'hla_match_a_high',
        'hepatic_severe', 'prior_tumor', 'hla_match_b_low', 'peptic_ulcer',
        'hla_match_a_low', 'rheum_issue', 'hla_match_b_high',
        'hepatic_mild', 'tce_div_match', 'hla_low_res_8', 'cardiac',
        'hla_match_drb1_high', 'pulm_moderate', 'hla_low_res_10'
    ]
    
    # Handle numerical features
    for column in numerical_columns:
        if df[column].isnull().any():
            # Create missing indicator
            df[f'{column}_missing'] = df[column].isnull().astype(int)
            
            # Convert to numeric, forcing non-numeric values to NaN
            df[column] = pd.to_numeric(df[column], errors='coerce')
            
            # Check if the column is actually numeric after conversion
            if df[column].dtype.kind in 'bifc':  # boolean, integer, float, complex
                if abs(df[column].skew()) > 1:
                    df[column] = df[column].fillna(df[column].median())
                else:
                    df[column] = df[column].fillna(df[column].mean())
            else:
                # If conversion to numeric failed, treat as categorical
                df[column] = df[column].fillna(df[column].mode()[0])
    
    # Handle categorical features that should use mode
    for column in mode_fill_columns:
        if df[column].isnull().any():
            df[column] = df[column].fillna(df[column].mode()[0])
    
    # Handle other categorical features
    for column in other_categorical_columns:
        if df[column].isnull().any():
            df[column] = df[column].fillna('unknown')
    
    # Drop unnecessary columns
    df = df.drop(columns=['ID'])
    if(test == 0):
        df.drop(columns=['efs_time'])
    return df

def clean_data_test(df):
    """
    Clean test data using the same approach as training data
    """
    return clean_data(df.copy(), 1)

df_clean = clean_data(train_df.copy())
df_clean
```

```python
def preprocess_for_model(df: pd.DataFrame, type: str) -> dict:
    """
    Preprocess data for model input
    Args:
        df: Input DataFrame
    Returns:
        dict: Features dictionary for model input
    """
    
    if type == 'train':
        # Clean data first
        df = clean_data(df.copy())
        df = df.drop(columns=['efs'])
    else:
        df = clean_data_test(df.copy())
    
    # # Create features dictionary with correct shapes
    features_dict = {}
    for name, values in df.items():
        if values.dtype == object:
            # Categorical features as strings
            features_dict[name] = values.astype(str).values.reshape(-1, 1)
        else:# Numeric features as float32
            features_dict[name] = values.astype('float32').values.reshape(-1, 1)
            
    return features_dict
```

```python
train_features = df_clean.copy()
train_labels = train_features.pop('efs')
```

## Encoding the categorical columns

```python
import tensorflow as tf
from tensorflow.keras import layers

import numpy as np
```

```python
inputs = {}

for name, column in train_features.items():
  dtype = column.dtype
  if dtype == object:
    dtype = tf.string
  else:
    dtype = tf.float32

  inputs[name] = tf.keras.Input(shape=(1,), name=name, dtype=dtype)

inputs
```

```python
numeric_inputs = {name:input for name,input in inputs.items()
                  if input.dtype==tf.float32}

x = layers.Concatenate()(list(numeric_inputs.values()))
norm = layers.Normalization()
norm.adapt(np.array(df_clean[numeric_inputs.keys()]))
all_numeric_inputs = norm(x)

all_numeric_inputs
```

```python
preprocessed_inputs = [all_numeric_inputs]
```

```python
for name, input in inputs.items():
    if input.dtype == tf.float32:
        continue

    # Convert all values to strings before finding unique values
    str_values = [str(x) for x in train_features[name]]
    unique_values = np.unique(str_values)
    
    # Create a preprocessing layer to convert to string
    string_casting = layers.Lambda(lambda x: tf.strings.as_string(x))
    
    # Setup lookup and encoding layers
    lookup = layers.StringLookup(vocabulary=unique_values)
    one_hot = layers.CategoryEncoding(num_tokens=lookup.vocabulary_size())

    # Apply the layers in sequence
    x = string_casting(input)
    x = lookup(x)
    x = one_hot(x)
    preprocessed_inputs.append(x)
```

```python
preprocessed_inputs_cat = layers.Concatenate()(preprocessed_inputs)

cibmtr_preprocessing = tf.keras.Model(inputs, preprocessed_inputs_cat)

# tf.keras.utils.plot_model(model = cibmtr_preprocessing , rankdir="LR", dpi=72, show_shapes=True)
```

```python
cibmtr_features_dict = {name: np.array(value) 
                         for name, value in train_features.items()}
```

```python
features_dict = {name:values[:1] for name, values in cibmtr_features_dict.items()}
cibmtr_preprocessing(features_dict)
```

```python
from sklearn import metrics

def cibmtr_model(preprocessing_head, inputs):
  body = tf.keras.Sequential([
    layers.Dense(64, activation='relu'),
    layers.Softmax(),    
    layers.Dense(1)
  ])

  preprocessed_inputs = preprocessing_head(inputs)
  result = body(preprocessed_inputs)
  model = tf.keras.Model(inputs, result)

  model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                optimizer=tf.keras.optimizers.Adam(), metrics=['accuracy'])
  return model

cibmtr_model = cibmtr_model(cibmtr_preprocessing, inputs)
```

```python
# First convert all values to proper tensors
train_features_dict = {}
for name, values in train_features.items():
    if name in ['donor_age', 'karnofsky_score', 'melphalan_dose', 'comorbidity_score', 
                'age_at_hct', 'year_hct', 'donor_age_missing', 'karnofsky_score_missing',
                'melphalan_dose_missing', 'comorbidity_score_missing']:
        # Handle numeric features
        train_features_dict[name] = tf.convert_to_tensor(values.values.astype(np.float32))
    else:
        # Handle categorical features
        train_features_dict[name] = tf.convert_to_tensor(values.astype(str).values)

# Fit the model
cibmtr_model.fit(x=train_features_dict, y=train_labels, epochs=100)
```
